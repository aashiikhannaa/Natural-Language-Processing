```python
from nltk.corpus import genesis

genesis.fileids()
```

```python
with genesis.open('english-kjv.txt') as f:
  data = f.read()
words = " ".join(data.split("\n"))

words
```

```python
#Tokenizing the words in the sentence, removing stop words and punctuations

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer

stopWords = set(stopwords.words('english'))
tokenizer = RegexpTokenizer(r'\w+')
words = tokenizer.tokenize(words)
words = " ".join(words)
tokens = word_tokenize(words)

wordsFiltered = []

for w in tokens:
  if w not in stopWords:
    wordsFiltered.append(w)

words = wordsFiltered

#words

```

4 (a)

Porter Stemmer

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
print(f"Root words using Porter Stemmmer\n")
for w in words:
  print(f"{w} => {ps.stem(w)}")
```

Snowball Stemmer

```python
from nltk.stem.snowball import SnowballStemmer

snow_stemmer = SnowballStemmer(language='english')
print(f"Root words using Snowball Stemmmer\n")
for w in words:
  print(f"{w} => {snow_stemmer.stem(w)}")
```

Lancaster Stemmer

```python
from nltk.stem.lancaster import LancasterStemmer
st = LancasterStemmer()
print(f"Root words using Lancaster Stemmmer\n")
for w in words:
  print(f"{w} => {st.stem(w)}")
```

Regexp stemmer

```python
from nltk.stem import RegexpStemmer
st = RegexpStemmer('ing$|s$|e$|out$', min=4) #removes substrings ing, s, e, out and get root word
print(f"Root words using Regexp Stemmmer\n")
for w in words:
  print(f"{w} => {st.stem(w)}")
```

WordNet Lemmatizer and others summed up

```python
from nltk import WordNetLemmatizer

ps = PorterStemmer()
snow_stemmer = SnowballStemmer(language='english')
st = LancasterStemmer()
regex = RegexpStemmer('ing$|s$|e$|out$', min=4)
lemmatizer = WordNetLemmatizer()

from prettytable import PrettyTable

incorrect = 0
total = len(words)

# Specify the Column Names while initializing the Table
myTable = PrettyTable(["Word", "Porter Stemmer", "Snowball Stemmer", "Lancaster Stemmer", "Regexp Stemmer", "Lemmatization (meaningful root word)"])

# Add rows
for w in words:
  myTable.add_row([f"{w}", f"{ps.stem(w)}", f"{snow_stemmer.stem(w)}", f"{st.stem(w)}", f"{regex.stem(w)}", f"{lemmatizer.lemmatize(w)}"])
  if ps.stem(w) != snow_stemmer.stem(w) or st.stem(w) != ps.stem(w) or ps.stem(w) != regex.stem(w):
    print(f"Different at {w}")
    incorrect += 1
```

```python
total, incorrect
correct = total-incorrect
print(f"Total: {total}\nIncorrect: {incorrect}\nCorrect: {correct}")

precision = (total-incorrect) / ((total-incorrect) + incorrect)
print(f"\nPrecision: {precision}")
```

```python
from transitions import Machine
```

```python
class MorphologicalGenerator:
  def __init__(self, lemma):
    self.lemma = lemma
    # Define the states
    self.states = ['lemma', 'plural', 'past', 'present']
    # Define the transitions
    self.transitions = [
    {'trigger': 'to_plural', 'source': 'lemma', 'dest': 'plural'},
    {'trigger': 'to_past', 'source': 'lemma', 'dest': 'past'},
    {'trigger': 'to_present', 'source': 'lemma', 'dest': 'present'},
    {'trigger': 'to_lemma', 'source': ['plural', 'past', 'present'], 'dest': 'lemma'}
    ]
    # Create a new state machine
    self.machine = Machine(model=self, states=self.states, transitions=self.transitions, initial='lemma')


  def generate_surface_form(self, tense, number):
    if tense == 'past':
      self.to_past()
    elif tense == 'present':
      self.to_present()
    elif tense == 'future':
      return self.lemma + ' will'

    if number == 'plural':
      self.to_plural()

    surface_form = self.lemma
    if self.state == 'past':
      if surface_form == "go":
        return "went"
      elif surface_form.endswith("y"):
        surface_form = surface_form[:-1]
        surface_form += "ied"
        return surface_form
      elif surface_form.endswith("e"):
        surface_form += "d"
        return surface_form
      elif surface_form == "sleep":
        return "slept"
      surface_form += 'ed'
    elif self.state == 'present':
      surface_form += 'ing'
    elif self.state == 'plural':
      surface_form += 's'
    return surface_form

# Define a lemma and its morphological features
lemma_word = 'work'
tense = 'past'
number = 'singular'

# Create a new instance of the morphological generator
generator = MorphologicalGenerator(lemma_word)
# Generate the surface form
surface_form = generator.generate_surface_form(tense, number)
# Print the surface form
print(surface_form)
```
